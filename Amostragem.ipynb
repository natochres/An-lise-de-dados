**Amostragem aleatória simples**
Em estatística, uma amostra aleatória simples é um subconjunto de indivíduos escolhidos de um conjunto maior no qual um subconjunto de indivíduos é escolhido aleatoriamente, todos com a mesma probabilidade. É um processo de selecionar uma amostra de forma aleatóri

```python
df_amostra_aleatoria_simples =  dataset.sample(n = 100)
```

```python
df_amostra_aleatoria_simples.shape
```

```python
df_amostra_aleatoria_simples.head(10)
```

```python
def amostragem_aleatoria_simples(dataset, amostras):
  return dataset.sample(n = amostras)
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(dataset, 100)
df_amostra_aleatoria_simples.shape
```

# **Amostragem sistemática**</p>
Na metodologia de levantamento, a amostragem sistemática é um método estatístico que envolve a seleção de elementos de uma base de amostragem ordenada. A forma mais comum de amostragem sistemática é um método de equiprobabilidade

```python
len(dataset) //100
```

```python
rd.seed(1)
rd.randint(0,325)
```

```python
rd.seed(1)
np.arange(rd.randint(0,325), len(dataset), step= 325 )
```

```python
def amostragem_sistematica(dataset, amostras):
  intervalo = len(dataset) //100
  rd.seed(1)
  inicio = rd.randint(0, intervalo)
  indices = np.arange(inicio, len(dataset), step = intervalo )
  amostragem_sistematica = dataset.iloc[indices]
  return amostragem_sistematica
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100)
df_amostra_sistematica.shape
```

```python
def amostragem_agrupamento (dataset, numero_grupos):
  intervalo = len(dataset) / numero_grupos

  grupos = []
  id_grupo = 0
  contagem = 0
  for _ in dataset.iterrows():
    grupos.append(id_grupo)
    contagem += 1
    if contagem > intervalo:
      contagem = 0
      id_grupo += 1
      
  dataset['grupo'] = grupos
  grupo_selecionado = rd.randint(0, numero_grupos)
  return dataset[dataset['grupo'] == grupo_selecionado]    
```

```python
df_amostra_agrupamento = amostragem_agrupamento(dataset, 100)
df_amostra_agrupamento.shape, df_amostra_agrupamento['grupo'].value_counts()
```
